```python
# Define job parameters
parameters_model = {
    'which_cats': 'singleCat',  # 'multipleCatsNoPersons', 'singleCat', 'multipleCats'
    'n_classes': 'all',  # 'all' or a specific number of classes
    'n_shuffle': 1,
    'n_iterations': 10,
    'n_folds': 10,
    'batch_size': 64
}

def run_model_training(params):
    import os
    script_command = (
        f"python train_model.py "
        f"{params['which_cats']} "
        f"{params['n_classes']} "
        f"{params['n_shuffle']} "
        f"{params['n_iterations']} "
        f"{params['n_folds']} "
        f"{params['batch_size']}"
    )
    print("Running command:", script_command)
    os.system(script_command)

# Run model training
run_model_training(parameters_model)
```
